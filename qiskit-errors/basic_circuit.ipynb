```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
import qiskit as qkit
import qiskit.visualization as qvis
```

```python
# prepare a qucntum circuit
qcirc = qkit.QuantumCircuit(3,3)

# add Hadamard gate
qcirc.h([0,2])

# add CNOT
qcirc.cx(0,1)

# final measurement
qcirc.measure_all()
```

```python
# see what is happening
qcirc.draw(output='mpl')
```

```python
# define the backend used to run the circuit
simulator = qkit.Aer.get_backend('qasm_simulator')
```

```python
# run the curcuit
job = qkit.execute(qcirc, simulator, shots=1000)
result = job.result()
```

```python
# analyse the result
qvis.plot_histogram(result.get_counts(qcirc))
```
